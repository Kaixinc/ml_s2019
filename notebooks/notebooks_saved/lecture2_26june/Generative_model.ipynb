# How to Emulate this Behavior in a Machine?
One approach is define:
**Hypothesis Space**
  Consider possible hypothesis $\mathcal{H}$ such as
  - Even number
  - odd number
  - power of 2
  - power of 2, plus 37
  - power of 2 except 32
  
  The subset of $\mathcal{H}$ consistent with data $\mathcal{D}$ is called **version space.**
  After seeing some example, some rules $h \in \mathcal{H}$ are consistent with
  samples. 
  
  How to combine them to predict $x_{test} \in C?$


## Some Observation About Version Space
- As we see more example version space shrinks
- How do you combine consistent rule in version space to predict if x_new ∈ C  
- Why we prefer one hypothesis more than other
     + After seeing D = {16, 8, 2, 64}, why did you choose the rule “powers of two” and not, say, “all even numbers”, or “powers of two except for 32”. They are all consistent with the evidence.
     
Can we explain why we choose power of 2 not "all even number?"(We want to **avoid suspicious coincidences** between power of 2 and even number. If even number was true concept how come we saw only even number with power of 2.)

  


## Bayesian explanation for this Choice
**Extension of a concept:** Set of all the number belonging to a concept.
  - Likelihood function.     
    Let assume example of sampled uniformly form the extension of concept $h$. 
    Then probability of independently sampling $N$ examples from $h$ with replacement is
    $$ p(\mathcal{D}|h) = \left[\frac{1}{size(h)}\right]^N$$
    
    $p(\mathcal{D}|h_{even} ) = 1/50$, after $\mathcal{D} =\{16 \}$
    and $p(\mathcal{D}|h_{two} ) = 1/6$, and this will increase after seeing
    $\mathcal{D} = {16, 8, 2, 64}$

    Model favors simplest(smallest) hypothesis consistent with the data( **Occam's razer**)
   . [William of Ockham](https://en.wikipedia.org/wiki/William_of_Ockham)



- Given D = {16, 8, 2, 64}, the concept h_1  =“powers of two except 32” is more likely than h =“powers of two”, 

How we resolve this issue?

Hypothesis powers of two except 32 seems bit unnatural? How to capture this intuition.

We can assign low prior probability to unnatural concepts(can bringing domain/ background knowledge). 

Which prior to choose(Subjectivity)? Why one prior is better than other?

Given hypothesis **prior** probability $h$ and some **evidence/likelihood** (data) $\mathcal{D}$,

how to update $p(h|\mathcal{D})$, called **posterior**

Note the need for the low prior on the unnatural concepts
 -  would have **overfit the data** and picked “powers of 2, except for 32”.)?

  When we have enough data posterior become peaked on single concept, called **MAP
  (Maximum a posteriori estimatie)** estimate.(**mode of the posterior**)
   \begin{align}
    \hat(h)^{MAP}& = \text{argmax}_h p(h|\mathcal{D})\\
    & = \text{argmax}_h p(\mathcal(D)|h) p(h) \\
    & = \text{argmax}_h [ \log p(\mathcal{D}|h) + \log p(h)] 
    \end{align}
    
   If more and more data(evidence come), eventually  $p(\mathcal(D)|h) p(h)$
   wins in the above expression.
   
   


## Maximum Likelihood Estimate(mle)

 $\hat(h)^{mle} = \text{argmax}_h p(\mathcal(D)|h)=  \text{argmax}_h \log p(\mathcal(D)|h)$

When data overwhelms prior **MAP estimate= MLE estimate**

If true hypothesis is in  hypothesis space $\mathcal{H}$ then eventually with more and more data we will converge to the true hypothesis.

If not then will converge to as close as possible to the truth.
