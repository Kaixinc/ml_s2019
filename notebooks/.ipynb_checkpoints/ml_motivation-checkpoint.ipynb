# Type of machine learning problem

## Primarily There are three types of machine learning problems

## Supervise (Main focus of this course)
Given a training set $\mathcal{D} = \{\mathbf{x_i}, y_i \}$, learn a rule/function $f$ from $\mathbf{x_i}$.
to y. $f: \mathbf{x} \rightarrow y$.
$\mathbf{x}$ is called feature, covariate, predictor, attribute and $y$ is called the response variable.

- if $y_i$ is a categorical variable {cat, dog, ...} then learning $f$ is known as classification problem
- if $y_i$ is real valued varaible, thne it known as regression problem in supervised setting 


 # Example
 - Linear/ Logistic regression
 - Navie Bayes
 - Support vector machine
 - Decision Tree
 - Neural networks

# Parametric vs non-parametric models
## Parametric: Model has a fixed number of parameters.
## non-parametric: Number of parameters grow with training data

## Parametric models for classification and regression
- Linear regression
$y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_m x_m + \epsilon$
data = signal + noise

$\epsilon \sim  \mathcal{N}{(\mu, \sigma^2)} = \frac{1}{\sqrt(2 \pi \sigma^2)} \exp(\frac{(x -\sigma)^2}{-2 \sigma ^2}) $

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Let's draw some sample from this distribution

```python
import numpy as np
mu, sigma = 0, 0.1 # mean and standard deviation

sample = np.random.normal(mu, sigma, 500)
print('mean and std of drawn sample is {}, {}'.format(np.mean(sample), np.std(sample)))
print(sample)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(10), range(10))
```

```python
count, bins, ignored = plt.hist(sample, 30, normed=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
                np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
          linewidth=2, color='r')
print(bins)
```
