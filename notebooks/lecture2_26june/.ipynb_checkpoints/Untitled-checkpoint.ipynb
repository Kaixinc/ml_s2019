```python
from scipy.stats import norm
import numpy as np
```

```python
x = np.linspace(-2, 2 , 100)
```

```python
print(x)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(x, norm.pdf(x),
        'r-', lw=5, alpha=0.6, label='norm pdf')
```

```python
plt.plot(x, norm.cdf(x), 'r-', lw=5, alpha=0.6, label='norm cdf')
```
